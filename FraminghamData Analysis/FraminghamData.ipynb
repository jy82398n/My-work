Two-Sample T-Test
In this case I will test whether the means of two independent data samples associated
In this case the hypothesis are:
H0: The two means are equal i.e Î¼(LDL|CVD=0) =Î¼(LDL|CVD=1)
Ha: The two means are not equal Î¼(LDL|CVD=0) â‰ Î¼(LDL|CVD=1)

Similarly for the HDLC
and taking  Î± = 0.05

```python
ds_0CVD = farmDF[farmDF.CVD == 0]
ds_1CVD = farmDF[farmDF.CVD == 1]
HDLC_mean = ds_0CVD["HDLC"].mean()
HDLC1_mean = ds_1CVD["HDLC"].mean()

print ("HDLC_mean (Mean):", HDLC_mean)
print ("HDLC1_mean (Mean):", HDLC1_mean)
```

```python
import scipy.stats as stats

stats.ttest_ind(ds_0CVD["HDLC"],ds_1CVD["HDLC"])
```

```python
LDLC_mean = ds_0CVD["LDLC"].mean()
LDLC1_mean = ds_1CVD["LDLC"].mean()

print ("LDLC_mean (Mean):", LDLC_mean)
print ("LDLC1_mean (Mean):", LDLC1_mean)
stats.ttest_ind(ds_0CVD["LDLC"],ds_1CVD["LDLC"])
```

As we can see that the means in both the cases are different 
Hence, we can see, the p-value is way smaller than 0.05 
and consequently p<Î± we reject the null hypothesis and say: 
"We have a statistically significant evidence to believe that HDLC and LDLC both are associated to CVD."

```python
cols=["HDLC", "LDLC","HYPERTEN","AGE","DIABP","SYSBP","DIAB"]
#cols=["HDLC", "LDLC"]
data=farmDF[cols].values
print(data)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
from sklearn import metrics

lr = LogisticRegression(solver='liblinear')
predicted = model_selection.cross_val_predict(lr,data, farmDF['CVD'])
print (metrics.accuracy_score(farmDF['CVD'], predicted))
```

```python
x = trainDF[cols].values
y = trainDF['CVD'].values

xTest = testDF[cols].values
yTest = testDF['CVD'].values

lr.fit(x,y)
print ("Coefficient: "+str(lr.coef_))
print ("Intercept: "+str(lr.intercept_))
print ("R-Squared: "+str(lr.score(xTest,yTest)))

predicted = lr.predict(xTest)
print (metrics.accuracy_score(yTest, predicted))
```
